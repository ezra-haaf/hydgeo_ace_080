## Flow to wells: steady state

With the boundary conditions $h(r=r_1)=h_1$ and $h(r=r_2)=h_2$ and a constant pumping rate Q in the well, the solution is:
$h_2-h_1=\frac{Q}{2\pi T}ln{\frac{r_2}{r_1}}$
